12. From the dataset, fit a linear model on the relationship between the temperature in the living room in Celsius (x = T2) and the temperature outside the building (y = T6). What is the R^2 value in two d.p.?

```python
# 12 Linear Regression Model of t2(Living Room Area) vs t6(Temprature Outside) in Degree Celsius

# Load the necessary libraries
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# t2  is the feature and t6 is the target
# Extract the features and target variable from the normalized dataset

features_df_t = df.drop(columns=['date','Appliances','lights', 'T1', 'RH_1', 'RH_2', 'T3', 'RH_3', 'T4', 'RH_4',
                                           'T5', 'RH_5', 'T6', 'RH_6', 'T7', 'RH_7', 'T8', 'RH_8', 'T9', 'RH_9',
                                           'T_out', 'Press_mm_hg', 'RH_out', 'Windspeed', 'Visibility',
                                           'Tdewpoint', 'rv1', 'rv2'])
target_t = df['T6']

scaler = MinMaxScaler()
normalised_df_t = pd.DataFrame(scaler.fit_transform(features_df_t), columns=features_df_t.columns)



# Split the data into training and testing sets
x_train_t2, x_test_t2, y_train_t6, y_test_t6 = train_test_split(features_df_t, target_t, test_size=0.3, random_state=42)

# Initialize and fit a Linear Regression Model
linear_model = LinearRegression()
linear_model.fit(x_train_t2, y_train_t6)

# Make predictions on the test set
predicted_values = linear_model.predict(x_test_t2)

# Calculate the R-squared score
r2_score_value = r2_score(y_test_t6, predicted_values)
rounded_r2_score = round(r2_score_value, 2)

# Print the R-squared score rounded to two decimal places
print("R^2 Score: ", rounded_r2_score)
```

13. Normalize the dataset using the MinMaxScaler after removing the following columns: [“date”, “lights”]. The target variable is “Appliances”. Use a 70-30 train-test set split with a random state of 42 (for reproducibility). Run a multiple linear regression using the training set and evaluate your model on the test set. Answer the following questions:

What is the Mean Absolute Error (in two decimal places)?

```python
# 2 Multiple Linear Regression of All features vs Application Energy Consumption

# Load the necessary libraries
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error


# Extract the features and target variable from the normalized dataset
scaler = MinMaxScaler()
normalised_df_app = df.drop(columns=['date','lights'])
normalised_df_app = pd.DataFrame(scaler.fit_transform(normalised_df_app), columns=normalised_df_app.columns)

target_app = normalised_df_app['Appliances']
features_df_app = normalised_df_app.drop(columns=['Appliances'], axis = 1)

# Split the data into training and testing sets
x_train_app, x_test_app, y_train_app, y_test_app = train_test_split(features_df_app, target_app, test_size=0.3, random_state=42)

# Initialize and fit a Linear Regression model
linear_model = LinearRegression()
linear_model.fit(x_train_app, y_train_app)

# Make predictions on the test set
predicted_values_app = linear_model.predict(x_test_app)

# Calculate the R-squared score
r2_score_value_app = r2_score(y_test_app, predicted_values_app)
rounded_r2_score_app = round(r2_score_value_app, 2)

# Calculate performance metrics
linear_regression_mae = mean_absolute_error(y_test_app, predicted_values_app)

# Print the R-squared score rounded to two decimal places
# print("R^2 Score: ", rounded_r2_score_app)
# print('\n')



print('MAE:', round(linear_regression_mae, 2))
```

14. What is the Residual Sum of Squares (in two decimal places)?

```python
# Residual Sum of Squares for (RSS) Appliances MUltivarte Linear Regression Model

rss = np.sum(np.square(y_test_app - predicted_values_app))
print('RSS:', round (rss, 2))
```

15. What is the Root Mean Squared Error (in three decimal places)

```python
# Root Mean Squared Error for Appliances MUltivarte Linear Regression Model
from sklearn.metrics import mean_squared_error
rmse = mean_squared_error(y_test_app, predicted_values_app)
rmse = round(np.sqrt(rmse), 3)
print('RMSE:', rmse)
```

17. Obtain the feature weights from your linear model above. Which features have the lowest and highest weights respectively?

```python
# Download Packages
from sklearn.linear_model import LinearRegression, Lasso, Ridge

# Initialize Models
linear_model = LinearRegression()
lasso_regression_model = Lasso(alpha=0.001)
ridge_regression_model = Ridge(alpha=0.4)

# Fit Models
linear_model.fit(x_train_app, y_train_app)
lasso_regression_model.fit(x_train_app, y_train_app)
ridge_regression_model.fit(x_train_app, y_train_app)

# Model, Features df and Column Names
def weights(model, features, col_name):
    # Retrieve feature weights (coefficients) from the model and sort them in ascending order
    if hasattr(model, 'coef_'):  # Check if the model has 'coef_' attribute
        weights_df = pd.Series(model.coef_, features.columns).sort_values()        
        # print('Lenght other Models:', len(weights_df))
        # print('Shape Other Models:', weights_df.shape)

    else:
        weights_df = pd.Series([0] * len(features.columns), features.columns).sort_values()  # Handle models without coefficients        
        # print('Linear Reg Len:', len(weights_df))        
        # print('Shape Lin_Reg:', weights_df.shape)

    # Assign Column names to the DataFrame
    weights_df = pd.DataFrame(weights_df).reset_index()
    weights_df.columns = ['Features', col_name]  # Assign correct column name

    # Round the values in the specified column to three decimal places
    weights_df[col_name] = weights_df[col_name].round(3)

    # Return the DataFrame with feature names and their corresponding weights
    return weights_df

linear_model_weights_df = weights(linear_model, x_train_app, 'Linear_Model_Weight')
ridge_weights_df = weights(ridge_regression_model, x_train_app, 'Ridge_Weight')
lasso_weights_df = weights(lasso_regression_model, x_train_app, 'Lasso_Weight')

final_weights_df = pd.merge(linear_model_weights_df, ridge_weights_df, on='Features', how = 'inner')
final_weights_df = pd.merge(final_weights_df, lasso_weights_df, on='Features', how = 'inner')
final_weights_df
```

```python
min_value = final_weights_df.loc[:, 'Linear_Model_Weight'].min()
max_value = final_weights_df.loc[:, 'Linear_Model_Weight'].max()

print('Min Val:', min_value)
print('Max Val:', max_value)
```

```python
# Retrieve the rows with the minimum and maximum values
min_rows = final_weights_df[final_weights_df['Linear_Model_Weight'] == min_value]
max_rows = final_weights_df[final_weights_df['Linear_Model_Weight'] == max_value]


print('Min Val:', min_rows)
print('Max Val:', max_rows)
```

```python
# Concatenate the rows into a single DataFrame
result_df = pd.concat([min_rows, max_rows])

print("Result DataFrame:")
print(result_df)
```
